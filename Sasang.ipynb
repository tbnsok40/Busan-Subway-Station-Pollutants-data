```python
import pandas as pd
nampo = pd.read_csv('사상.csv', encoding = 'cp949', index_col = 0)
nampo
```

```python
nampo['PM2.5']=pd.to_numeric(nampo['PM2.5'], errors='coerce')
print(nampo['PM2.5'].mean())
print(nampo['PM2.5'].std())
nampo['PM10']=pd.to_numeric(nampo['PM10'], errors='coerce')
print(nampo['PM10'].mean())
print(nampo['PM10'].std())
```
