```python
import pandas as pd
from pandas import Series, DataFrame
jan = pd.read_csv('C:/지하철/20140101.csv')
feb = pd.read_csv('C:/지하철/20140201.csv')
mar = pd.read_csv('C:/지하철/20140301.csv')
apr = pd.read_csv('C:/지하철/20140401.csv')
may = pd.read_csv('C:/지하철/20140501.csv')
jun = pd.read_csv('C:/지하철/20140601.csv')
jul = pd.read_csv('C:/지하철/20140701.csv')
aug = pd.read_csv('C:/지하철/20140801.csv')
sep = pd.read_csv('C:/지하철/20140901.csv')
oct = pd.read_csv('C:/지하철/20141001.csv')
nov = pd.read_csv('C:/지하철/20141101.csv')
dec = pd.read_csv('C:/지하철/20141201.csv')
```

# 서면 승강장 평균

## 평균(31일)

```python

def avg(month):
    filter1 = month[['최저','최고','평균']]
    평균 = filter1['평균']

    t=1
    x=0
    서면=[]

    while t<=35:
        서면.append(평균[36+x])
        x += 38
        t += 1

    df = DataFrame({'month':서면}, 
    index=['서면1 no2','co2','pm2.5','pm10','co','서면2 no2','co2','pm2.5','pm10','co','남포 no2','fad','co2','pm10','co','서면1 no2','co2','pm2.5','pm10','co','서면2(대) no2','fad','co2','pm2.5','pm10','co','사상(대) no2','fad','co2','pm10','co','미남 no2','co2','pm10','co'])
    return df


jan_mean = avg(jan)
mar_mean = avg(mar)
may_mean = avg(may)
jul_mean = avg(jul)
aug_mean = avg(aug)
oct_mean = avg(oct)
dec_mean = avg(dec)

```

## 평균 (30일)

```python
def avg(month):
    filter1 = month[['최저','최고','평균']]
    평균 = filter1['평균']

    t=1
    x=0
    서면=[]

    while t<=35:
        서면.append(평균[35+x])
        x += 37
        t += 1

    df = DataFrame({'month':서면}, 
    index=['서면1 no2','co2','pm2.5','pm10','co','서면2 no2','co2','pm2.5','pm10','co','남포 no2','fad','co2','pm10','co','서면1 no2','co2','pm2.5','pm10','co','서면2(대) no2','fad','co2','pm2.5','pm10','co','사상(대) no2','fad','co2','pm10','co','미남 no2','co2','pm10','co'])
    return df


apr_mean = avg(apr)
jun_mean = avg(jun)
sep_mean = avg(sep)
nov_mean = avg(nov)
```

## 평균(29일)

```python
def avg(month):
    filter1 = month[['최저','최고','평균']]
    평균 = filter1['평균']

    t=1
    x=0
    서면=[]

    while t<=35:
        서면.append(평균[33+x])
        x += 35
        t += 1

    df = DataFrame({'month':서면}, 
    index=['서면1 no2','co2','pm2.5','pm10','co','서면2 no2','co2','pm2.5','pm10','co','남포 no2','fad','co2','pm10','co','서면1 no2','co2','pm2.5','pm10','co','서면2(대) no2','fad','co2','pm2.5','pm10','co','사상(대) no2','fad','co2','pm10','co','미남 no2','co2','pm10','co'])
    return df
feb_mean = avg(feb)
```

## 평균 합병

```python
sm = pd.concat([jan_mean,feb_mean,mar_mean,apr_mean,may_mean,jun_mean,jul_mean,aug_mean,sep_mean,oct_mean,nov_mean,dec_mean], axis =1)
sm
```

## 연평균

```python
mean_final = sm.mean(axis=1)
mean_final
mean_final.to_csv('C:/지하철/mean_finalll.csv', encoding = 'utf-8-sig')
```
