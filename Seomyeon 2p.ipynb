```python
def automate_alltime(df):
    fig, (ax1) = plt.subplots(nrows=1, ncols=1)
    fig.set_size_inches(18,10)
    sns.boxplot(x="year", y="PM10 (μg/m³)", data=df, ax= ax1, color = 'w')
    ax1.set(ylim=(-10, 300)) ## 750위로 9개 점(5개년에 걸쳐)
    plt.show()

    fig, (ax1) = plt.subplots(nrows=1, ncols=1)
    fig.set_size_inches(18,10)
    sns.boxplot(x="year", y="PM2.5 (μg/m³)", data=df, ax= ax1, color = 'w')
    ax1.set(ylim=(-10, 200)) ## 800위로 12개 점 (5개년에 걸쳐)

    plt.show()
    return
```

```python
def automate_separate(df):
    
    df['OP'] = 'OP'
    df.loc[(df['hour']<=5)&(df['hour']>=1), 'OP'] = 'NON'
    
    fig, (ax1) = plt.subplots(nrows=1, ncols=1)
    fig.set_size_inches(18,10)

    sns.boxplot(x="year", y="PM10 (μg/m³)", data=nampo, ax= ax1, hue = nampo['OP'])
    ax1.set(ylim=(-10, 300)) ## 800위로 12개 점 (5개년에 걸쳐)

    plt.show()
    
    
    fig, (ax1) = plt.subplots(nrows=1, ncols=1)
    fig.set_size_inches(18,10)
    sns.boxplot(x="year", y="PM2.5 (μg/m³)", data=nampo, ax= ax1, hue = nampo['OP'])
    ax1.set(ylim=(-10, 200)) ## 800위로 12개 점 (5개년에 걸쳐)

    plt.show()
    
 
    return
```

```python
automate_separate(nampo)
```
