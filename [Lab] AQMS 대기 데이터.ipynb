### dataframe의 인덱스 확인하면서, time frame이 들어갈 개수 계산해야 오류 안난다

```python
# pd.date_range(start = "2014-1-1", end='2020-1-31-24', freq = 'H')
# time = pd.date_range("2014-1-1-1", "2020-1-31-24", freq='H')
# time
온천['time']=pd.date_range(start='2014-1-1-01', end='2020-2-1-00', freq = 'H')
온천
```

```python

온천["time-year"] = 온천["time"].dt.year

온천["time-month"] = 온천["time"].dt.month

온천["time-day"] = 온천["time"].dt.day

온천["time-hour"] = 온천["time"].dt.hour
```

```python
온천
```

```python
온천14 = 온천.loc[온천['time-year']==2014]
```

```python
온천_non= (온천14['time-hour']<=23) & (온천14['time-hour']>=5)
온천_non
```

```python
온천non = 온천14.loc[온천_non]
온천non
```

```python
온천non['CO']=온천non['CO'].apply(pd.to_numeric, errors='coerce')
온천non['NO₂']=온천non['NO₂'].apply(pd.to_numeric, errors='coerce')
온천non['PM10']=온천non['PM10'].apply(pd.to_numeric, errors='coerce')
```

```python
figure, (ax1,ax2,ax3) = plt.subplots(nrows=3, ncols=1)
figure.set_size_inches(18,30)
sns.lineplot(data=온천non, x='time-hour', y = 'CO', ax = ax1)
sns.lineplot(data=온천non, x='time-hour', y = 'NO₂', ax = ax2)
sns.lineplot(data=온천non, x='time-hour', y = 'PM10', ax = ax3)
```
