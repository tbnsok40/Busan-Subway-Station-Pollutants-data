```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

남포 = pd.read_csv("남포.csv", encoding = 'cp949')
남포 = 남포.apply(pd.to_numeric, errors='coerce')
남포
```

```python
figure, (ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(nrows=6, ncols=1)
sns.boxplot(x="datetime-year", y="CO", data=남포, ax= ax1)
sns.boxplot(x="datetime-year", y="CO2", data=남포, ax= ax2)
sns.boxplot(x="datetime-year", y="NO2", data=남포, ax= ax3)
sns.boxplot(x="datetime-year", y="PM2.5", data=남포, ax= ax4)
sns.boxplot(x="datetime-year", y="PM10", data=남포, ax= ax5)
sns.boxplot(x="datetime-year", y="HCHO", data=남포, ax= ax6)
figure.set_size_inches(15,60)
plt.show()
```
