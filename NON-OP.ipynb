```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
non_op = pd.read_csv('C:/codeit/non_op.csv')
non_op
```

```python
non_op=non_op[['Unnamed: 1','2014','2015','2016','2017','2018','2019']]
op_co = non_op.loc[0:5, 'Unnamed: 1': '2019']
op_co = op_co.set_index('Unnamed: 1')
op_co = op_co.T
op_co.plot()
plt.title("CO")
plt.xlabel("Year")
plt.ylabel("Concenturation")
plt.rcParams["figure.figsize"] = (20,8)
ax1 = plt.show()
ax1
```
