```python
def automation(df,year):
    newdf = df.loc[df['년']==year]
    newdf = newdf[['CO','NO₂','PM10','PM2.5']]
    newdf = newdf.apply(pd.to_numeric, errors='coerce')
    return  newdf[['CO','NO₂','PM10','PM2.5']].mean()

print(automation(전포,2014))
print(automation(전포,2015))
print(automation(전포,2016))
print(automation(전포,2017))
print(automation(전포,2018))
print(automation(전포,2019))
print(automation(전포,2020))
```

```python
def automate(df):
    df['OP'] = 0
    df.loc[(df['시']<=23)&(df['시']>=5), 'OP'] = 1
    df = df[['년','월','일','시','CO','NO₂','PM10','PM2.5','OP']]
    df = df.apply(pd.to_numeric, errors='coerce')
    figure, (ax1,ax2,ax3,ax4) = plt.subplots(nrows=4, ncols=1)
    sns.boxplot(x="년", y='CO', data=df, ax= ax1, hue='OP')
    sns.boxplot(x="년", y='NO₂', data=df, ax= ax2, hue='OP')
    sns.boxplot(x="년", y='PM10', data=df, ax= ax3, hue='OP')
    sns.boxplot(x="년", y='PM2.5', data=df, ax= ax4, hue='OP')
    figure.set_size_inches(25,40)
    return plt.show()
```
